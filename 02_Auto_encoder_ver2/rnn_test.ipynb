```python
input_size = len(char2idx) # 37
hidden_size = 128
output_size = len(char2idx) # 37
batch_size = 1
sequence_length = len(x_data)
learning_rate = 0.001

x_data = np.reshape(x_data, [batch_size, len(x_data)])
y_data = np.reshape(y_data, [batch_size, len(y_data)])

print (np.shape(x_data))
print (np.shape(y_data))
```

### Build a RNN model

```python
X = tf.placeholder(tf.int32, [None, sequence_length], name = 'x_data')
Y = tf.placeholder(tf.int32, [None, sequence_length], name = 'y_data')
# one_hot : [sequence_length, input_size]
x_one_hot = tf.one_hot(X, input_size) 

with tf.variable_scope("RNN"):
    
    cell = tf.contrib.rnn.BasicLSTMCell(num_units=hidden_size,
                                                       state_is_tuple=True)

    # defining initial state
    initial_state = cell.zero_state(batch_size, dtype=tf.float32)

    outputs, state = tf.nn.dynamic_rnn(cell=cell,
                                                       inputs=x_one_hot,
                                                       initial_state=initial_state)
    # output size : [batch_size, seqence_length, hidden_size]
    X_for_fc = tf.reshape(outputs, [-1, hidden_size])
    outputs = layers.fully_connected(inputs=X_for_fc,
                                                   num_outputs=output_size,
                                                   activation_fn=None)
    # reshape outputs for sequence_loss
    outputs = tf.reshape(outputs, [batch_size, sequence_length, output_size])
    
    FC_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope='RNN')

with tf.variable_scope("loss"):
    weights = tf.ones([batch_size, sequence_length]) 
    sequence_loss = tf.contrib.seq2seq.sequence_loss(logits=outputs,
                                                                                 targets=Y,
                                                                                 weights=weights)
    loss = tf.reduce_mean(sequence_loss)
    train = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(loss, var_list=FC_vars)
    tf.summary.scalar("loss", loss)
    
prediction = tf.argmax(outputs, axis=2)
```

### Train the model

```python
step = 1000
with tf.Session() as sess:
    writer = tf.summary.FileWriter("./graphs/rnn", sess.graph)
    summary_op = tf.summary.merge_all()
    sess.run(tf.global_variables_initializer())
    
    print("y: ", train_melody[1: ])
    for i in range(step):
        loss_val, _ = sess.run([loss, train], feed_dict={X:x_data, Y:y_data})
        result = sess.run(prediction, feed_dict={X:x_data})
        
        summary = sess.run(summary_op, feed_dict={X:x_data, Y:y_data})
        
        result_str = [idx2char[c] for c in np.squeeze(result)]
        error = np.array(result_str) - np.array(train_melody[1:])
        
        print( "{:4d}  loss: {:.5f}".format(i, loss_val), "    Prediction: ", result_str, "    error: ", error)
        
        writer.add_summary(summary, i)
```
